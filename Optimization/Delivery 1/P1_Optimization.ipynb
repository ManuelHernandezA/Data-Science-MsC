#1. One dimensional case

Given the following function:
$$
f(x)=x^3-2x+2
$$

**1. Plot this function within the range $x \in [−2, 2]$.**

```python
import matplotlib.pyplot as plt
import numpy as np

# _Function to plot
X = np.arange(-2, 2, 0.25)
Y = X**3-2*X+2;

# Plot
fig, ax = plt.subplots()
line, = plt.plot(X, Y, label='$f(x)$')
ax.legend(loc='upper left', frameon=False)
plt.show()
```

**2. Compute analytically the points $x^*$ that satisfy $f'(x) = 0$**.

First, we compute analytically the first derivative of the function, which is given by:

$$
f'(x)=3x^2-2 \\
$$

We now find the point $x$ were $f'(x) = 0$:

$$
f'(x)=0 \\
3x^2 - 2 = 0\\
x^2=2/3 \\
$$

From the above, we get then these two solutions $x^*$ that satisfy $f'(x) = 0$:

$$
x^*=\pm\sqrt{2/3} \\
$$

**3. Check which points $x^*$ are maxima or minima.**

First we find the second derivative of the function, which is given by:

$$
f''(x)=6x
$$

We are now going to evaluate the points $x^*=\pm\sqrt{2/3}$ obtained in the previous exercise.
If we consider the expression $d^2f''(x)$ of the 2nd order Taylor expansion, it can be seen that,
for every $d$ such that $d\neq 0$, and with $x=\sqrt{2/3}$, it will always hold that $d^2f''(x) > 0$.

Similarly, having $x=-\sqrt{2/3}$, and for every $d$ such that $d\neq 0$, it will hold that $d^2f''(x) < 0$.

From the above, we can conclude that, $x=\sqrt{2/3}$ is a minima of the function and $x=-\sqrt{2/3}$ is a maxima of the function.

**4. Plot $f''(x)$ within the range $x \in [−2, 2]$.**

```python
# Function to plot
X = np.arange(-2, 2, 0.25)
Y = 6*X;

# Plot
fig, ax = plt.subplots()
line, = plt.plot(X, Y, label='$f\'\'(x)$')
ax.legend(loc='upper left', frameon=False)
plt.show()
```

For $x \in (0, 2]$ we can see that $f''(x) \geq 0$, so we can approximate $f(x)$ as a 2nd order Taylor expansion at $x$ such as (where $a$ is the point of approximation):
$$
x^3-2x+2+\frac{3x^2-2}{1!}(x-a)+\frac{6x}{2!}(x-a)
$$

And for $x \in [-2, 0)$, the function is concave, so its not interesting for finding minima.

# 2. Two dimensional case

Let us consider the function:

$$
f(\mathbf{x}) = x_1^2 + x_2^2
$$

**1. Plot this function.**

With the following code we plot the aforementioned function:

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np


fig = plt.figure()
ax = plt.axes(projection='3d')

# Make data.
X = np.arange(-2, 2, 0.1)
Y = np.arange(-2, 2, 0.1)
X, Y = np.meshgrid(X, Y)
Z = (X**2) + (Y**2)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(0.0, 7.01)
ax.set_zlim(0, 10)
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$x_1^2 + x_2^2$')
ax.view_init(20, 35) # rotate view

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=10)

plt.title('$x_1^2 + x_2^2$')
plt.show()
```

We can clearly see that the function has a minimum at $(x_1, x_2) = (0,0)$.


**2. Analitically compute the gradient of the function $\nabla f(\mathbf{x})$ and compute $\mathbf{x^*}$ where
$\nabla f(\mathbf{x^*}) = \mathbf{0}$:**

The gradient $\nabla f(\mathbf{x})$ of the function is given by the following expression:

$$
\nabla f(\mathbf{x}) = (2x_1, 2x_2)
$$

It is straighforward to see that the point $\mathbf{x^*}$ where $\nabla f(\mathbf{x^*})=0$ is given by $\mathbf{x^*} = (0, 0)$.


**3. Compute the Hessian matrix**

$$
\nabla^2 f(\mathbf{x}) =
\begin{pmatrix}
\frac{\partial^{2}f}{\partial x^2} & \frac{\partial^{2}f}{\partial xy} \\
\frac{\partial^{2}f}{\partial xy} & \frac{\partial^{2}f}{\partial y^2}
\end{pmatrix}
=
\begin{pmatrix}
2 & 0 \\
0 & 2
\end{pmatrix}
$$

Since, in this case, the Hessian is a constant matrix, clearly we have $\nabla^2 f(\mathbf{x}) = \nabla^2 f(\mathbf{x^*})$.

**Which are the eigenvalues of the Hessian matrix?
Do we have a minimum or a maximum at $x^*$?**

With the following operations we extract the eigenvalues of $\nabla^2 f(\mathbf{x^*})$

```python
H = np.array([[2,0],[0,2]])
eigenvalues, eigenvectors = np.linalg.eig(H)
for i in range(eigenvalues.shape[0]):
  print(f"eigenvalue  {i}:", eigenvalues[i])
  print(f"eigenvector {i}:", eigenvectors[:,i])
  print()
```

We get the eigenvectors
$v_1 = (1, 0)$, $v_2 = (0, 1)$ and their corresponding eigenvalues $\lambda_1 = 2, \lambda_2 = 2$
It is straighforward to check that the definition of eigenvalue and eigenvector holds for the values obtained: $\nabla^2 f(\mathbf{x^*})v = \lambda v$.

Notice that, in this case the eigenvalues of the Hessian are strictly positive, which means that the following is satisfied: $\mathbf{d}^T\nabla^2 f(\mathbf{x^*}) \mathbf{d} > 0$ with $\mathbf{d} \neq \mathbf{0}$. Therefore, we can conclude that we have a minimum at $\mathbf{x^*}$.


**4. What happens if some eigenvalues are positive and some
negative? What happens if the eigenvalue is zero?**

Let us consider the following functions:

$$
f_A(\mathbf{x}) = -x_1^2 - x_2^2 \\
f_B(\mathbf{x}) = x_1^2 - x_2^2 \\
f_C(\mathbf{x}) = x_1^2 \\
$$

**(a) Perform a plot of the function. At which point $\mathbf{x^∗}$ is the gradient zero?**

Here we provide the countour plot of the functions:

```python
def countour_plot(fig, ax, function, title="", xmin=-3.0, xmax=3.0, ymin=-3.0, ymax=3.0, steps=100):
  xlist = np.linspace(xmin, xmax, steps)
  ylist = np.linspace(ymin, ymax, steps)
  X, Y = np.meshgrid(xlist, ylist)
  Z = function(X, Y)
  cp = ax.contourf(X, Y, Z)
  fig.colorbar(cp) # Add a colorbar to a plot
  ax.set_title(title)
  ax.set_xlabel('x_1')
  ax.set_ylabel('x_2')

def fA(X1, X2):
  return -X1**2 -X2**2

def fB(X1, X2):
  return X1**2 - X2**2

def fC(X1, X2):
  return X1**2

#fig,ax=plt.subplots(1,1) # uncomment for 1-row plot
plt.rcParams['figure.figsize'] = [12, 3] # resize notebook plots
fig, (ax1, ax2, ax3) = plt.subplots(1,3)
countour_plot(fig, ax1, fA, title="$-x_1^2-x_2^2$")
countour_plot(fig, ax2, fB, title="$x_1^2-x_2^2$")
countour_plot(fig, ax3, fC, title="$x_1²$")
plt.show()
```

The gradients of the functions are provided below:

$$
\nabla f_A(\mathbf{x}) = (-2x_1, -2x_2) \\
\nabla f_B(\mathbf{x}) = (2x_1, -2x_2) \\
\nabla f_C(\mathbf{x}) = (2x_1, 0) \\
$$

From both the countour plots and gradients we can obtain the values $\mathbf{x^*}$ where the gradient is zero:

$$
\nabla f_A(\mathbf{x^*}) = \mathbf{0}
\rightarrow \mathbf{x^*}=(0, 0)\\
\nabla f_B(\mathbf{x^*}) = \mathbf{0}
\rightarrow \mathbf{x^*}=(0, 0) \\
\nabla f_C(\mathbf{x^*}) = \mathbf{0}
\text{  when  } x_1^*=0 \\
$$

**(b) At the points where the gradient is zero, what kind of information is giving us the Hessian
matrix?**

We now compute the Hessian matrix of the functions, which are given by:

$$
\nabla^2 f_A(\mathbf{x}) =
\begin{pmatrix}
-2 & 0 \\
0 & -2
\end{pmatrix} \\
\nabla^2 f_B(\mathbf{x}) =
\begin{pmatrix}
2 & 0 \\
0 & -2
\end{pmatrix} \\
\nabla^2 f_C(\mathbf{x}) =
\begin{pmatrix}
2 & 0 \\
0 & 0
\end{pmatrix} \\
$$

Finally, we are ready to compute the eigenvalues and eigenvectors of the matrix:

```python
import numpy as np

def compute_and_print_spec(H, function_name=""):
  eigenvalues, eigenvectors = np.linalg.eig(H)
  for i in range(eigenvalues.shape[0]):
    print(f"eigenvalue  {i} of {function_name} :", eigenvalues[i])
    print(f"eigenvector {i} of {function_name}:", eigenvectors[:,i])
    print()

compute_and_print_spec(np.array([[-2,0],[0,-2]]), function_name="Hessian f_A")
compute_and_print_spec(np.array([[2,0],[0,-2]]), function_name="Hessian f_B")
compute_and_print_spec(np.array([[2,0],[0,0]]), function_name="Hessian f_C")
```

From the above operations we obtain the following eigenvalues:

$$
\lambda_{A} = (-2, -2) \\
\lambda_{B} = (2, -2) \\
\lambda_{C} = (2, 0) \\
$$

where $\lambda_{i}$ is used to denote the eigenvalues of the Hessian of function $f_i(\mathbf{x})$ (i.e. the eigenvalues of the matrix $\nabla^2 f_i(\mathbf{x})$).

From the above eigenvalues, we can see negative eigenvalues indicate a concave function, positive eigenvalues indicate a convex function, and that both positive and negatives eigenvalues (e.g. $\lambda_{B} = (2, -2)$), indicate a saddle point in the function.

# 2.2 A two dimensional function with multiple minima

Given the function:

$$
f(x_1, x_2) = x_1^2(4 - 2.1 x_1^2 + \frac{1}{3}x_1^4) + x_1 x_2 + x_2^2(-4 + 4x_2^2)
$$

Here we plot the function within the range $x_1 \in [−2, 2]$ and $x_2 \in [−1, 1]$.

```python
def function(X, Y):
  return ((X**2) * (4.0 - (2.1 * (X**2)) + ((1.0/3.0)*(X**4)))) \
    + (X*Y) \
    + ((Y**2) * (-4.0 + (4.0 * (Y**2))))
```

```python
plt.rcParams['figure.figsize'] = [11, 4] # resize notebook plots
fig = plt.figure()

ax1 = fig.add_subplot(1, 2, 1, projection='3d')

# Make data.
X = np.arange(-2, 2, 0.1)
Y = np.arange(-1, 1, 0.1)
X, Y = np.meshgrid(X, Y)
Z = function(X, Y)

# Plot the surface.
surf = ax1.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax1.set_zlim(0, 8)
ax1.set_xlabel('$x_1$')
ax1.set_ylabel('$x_2$')
ax1.set_zlabel('$x_1^2 + x_2^2$')
ax1.view_init(20, 250) # rotate view

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=10)

ax2 = fig.add_subplot(1, 2, 2)
countour_plot(fig, ax2, function, title="$f(x_1, x_2)$", xmin=-2.0, xmax=2.0, ymin=-1.0, ymax=1.0, steps=200)

plt.show()
```

From the plot we can estimate the minima to be at $(0, 0.75)$ and $(0, -0.75)$ and the maxima to be at $(-2, -1)$ and $(2, 1)$.

**2. Analytically compute the gradient $\nabla f(\mathbf{x})$.**

The gradient of $f(x_1, x_2)$ is given by:

$$
\nabla f(x_1, x_2) = [x_2 + 2x_1^5 - 8.4x_1^3 + 8x_1,\ \ x_1 + 16x_2^3 - 8x_2]
$$

**3. Numerically compute an approximation of points $\mathbf{x^*}$ at which $\nabla f (\mathbf{x}^*)=0$**.

To approximate the points $\mathbf{x^*}$, we evaluate $\|\nabla f(\mathbf{x})\|^2$, using a grid built with a step of 0.005 in the range $x_1 \in [-2, 2]$ and $x_2 \in [-1, 1]$. The approximated points $\mathbf{\tilde{x}}$ will be the ones where $\|\nabla f(\mathbf{x})\|^2$ is strictly smaller than their neighbours. For simplicity, border points of the grid are ignored.

```python
step = 0.005
x1min, x1max = -2.0, 2.0
x2min, x2max = -1.0, 1.0
```

```python
def grad(x1, x2):
  return [ \
      x2 + (2.0*(x1**5)) - (8.4*(x1**3) + (8.0*x1)),
      x1 + (16.0*(x2**3)) - (8.0*x2)
  ]

def norm(g1, g2):
  return np.sqrt(g1**2+g2**2)
```

```python
rows = int((x1max - x1min) / step)
cols = int((x2max - x2min) / step)
sg = np.zeros((rows, cols))

for i in range(sg.shape[0]):
  for j in range(sg.shape[1]):
    g = grad(x1min + (i * step), x2min + (j * step))
    sg[i, j] = norm(g[0], g[1])**2
```

```python
def neighbours(grid, i, j):
    rows = np.array([-1, -1, -1,  0,  0,  1,  1,  1])
    cols = np.array([-1,  0,  1, -1,  1, -1,  0,  1])
    return grid[rows+i,cols+j]

min_grad = []
for i in range(1, sg.shape[0] - 1):
  for j in range(1, sg.shape[1] - 1):
    if np.all(sg[i][j] < neighbours(sg, i, j)):
      x1 = x1min + (i * step)
      x2 = x2min + (j * step)
      min_grad.append((x1, x2, sg[i][j]))
```

Here we calculate the values of $\mathbf{\tilde{x}}$, this is, our approximation of $\mathbf{x^*}$, along with the value of $\|\nabla f(\mathbf{x})\|^2$ at these points:

```python
for x1, x2, gr in min_grad:
  print(f"(x_1, x_2) = ({x1: .2f}, {x2: .2f}) with grad = {gr:.2f}")
```

**4. Compute the analytical expression of the Hessian of $f(x_1, x_2)$ and evaluate it at the values
$\mathbf{\tilde{x}}$ you have found. What kind of information is giving you the Hessian?**

The analytical expression of the Hessian is given by:

$$
\nabla^2 f(x_1, x_2) =
\begin{pmatrix}
10 x_1^4 - 25.2 x_1^2 + 8& 1 \\
1 & 48 x_2^2 - 8
\end{pmatrix} \\
$$

We now compute the values of the Hessian matrix $\nabla^2 f(x_1, x_2)$ for each of the points $\mathbf{\tilde{x}}$ found previously. For each point $\mathbf{\tilde{x}}$ we also compute the eigenvalues and eigenvectors of $\nabla^2 f(\mathbf{\tilde{x}})$.

```python
def hessian_calc(x1,x2):
  return np.array([[10*(x1**4)-25.2*(x1**2)+8, 1], [1, 48*(x2**2)-8]])

for x1, x2, _ in min_grad:
  print("---------------------------------------------------------------")
  print(f"Hessian matrix at ({x1:.2f}, {x2:.2f})")
  H = hessian_calc(x1,x2)
  for i in range(H.shape[0]):
    for j in range(H.shape[1]):
      print(f"  {H[i,j]:.2f}", end="\t")
    print()
  print()
  compute_and_print_spec(H,"the given function")
```

Given the computed eigenvalues $\lambda_1,\lambda_2$ of the Hessian matrix for the critical points $\mathbf{\tilde{x}}$ we can conclude that:

*   $x_1, x_2 = (-0.08, -0.70)$ with $\lambda_1,\lambda_2 = (7.69, 15.65)$ is a minima since both eigenvalues are positive definite.
*   $x_1, x_2 = (0.00, 0.00)$ with $\lambda_1,\lambda_2 = (8.06, -8.06)$ is a saddle point since the eigenvalues are positive and negative.
*   $x_1, x_2 = (0.08, 0.70)$ with $\lambda_1,\lambda_2 = (7.69, 15.65)$ is a minima since both eigenvalues are positive definite.
